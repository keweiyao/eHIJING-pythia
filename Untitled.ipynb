```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
@np.vectorize
def phase(l,k,Q):
    l2 = l**2
    k2 = k**2
    def df(x):
        cos = np.cos(x)
        L2 = l2 + k2 + 2*k*l*cos
        if L2>Q**2:
            return 0.
        return (k*l*cos+k2)/L2/2/np.pi
    return quad(df, 0, 2*np.pi)[0]
```

```python
Q = 4
l = 1
k = np.exp(np.linspace(-2,3,100))
fig, axes = plt.subplots(5,1,figsize=(6,8), sharex=True)
for ax, l, c in zip(axes, [0.5, 1, 2, 4,8],'rgbck'):
    ax.plot(k, phase(l,k,Q)/(k**2+.5**2), color=c, label=r'$\int d\phi$ with $|l_\perp\'+k_\perp|<Q$')
    ax.plot(k, (l<k)/(k**2+.5**2),'--', color=c, label=r'$\Theta()$ approximation')
    ax.semilogx()
    ax.annotate(r'$l_\perp\'={}$ GeV'.format(l), xy=(.01,.7), xycoords="axes fraction", fontsize=12)
axes[2].set_ylabel(r"$\int(...)d\phi_k/(k^2+0.25)$", fontsize=14)
axes[-1].set_xlabel(r"$k$ [GeV]")
axes[0].set_title(r"$Q = 4$ GeV", fontsize=14)
axes[0].legend()
plt.subplots_adjust(hspace=0)
plt.savefig("Demo-phi.png",dpi=300)
```
